### visualization

#### Univariate Analysis

```python
#checking distribution of dependent Feature
sns.histplot(df["Selling Price"])
```

### Insights
- its right skewed
- majority of observations are concentrated on the lower end of  price

```python
#checking for the outlier
sns.boxplot(df["Selling Price"])
```

### Insights
- Avergae value of phones in the dataset are near 10k to 30k price range (budget segment)
- while the Maximum values is near 60k (premimum segment)
- here outliers can be ultra premium models of brands (ultra premium segment)

```python
sns.set_theme(style="whitegrid")
colors = sns.color_palette('pastel')

labels =['SAMSUNG ', 'Apple ', 'realme ', 'OPPO ', 'Nokia', 'Xiaomi', 'GIONEE',
       'Lenovo', 'ASUS', 'Motorola', 'LG', 'vivo', 'Infinix',
       'POCO', 'HTC ', 'Google Pixel ', 'IQOO ']
plt.pie(df["Brand"].value_counts(), labels=labels, colors=colors, autopct='%.0f%%')
```

### Insights
- pie digram show that samsung have most model in the dataset (25%)
- followed by realme , Oppo, Nokia and Xiaomi (13%,11%,9%,7%,6%) 
- while brands like Gionee , Lenovo , Asus , Motorola have 4% 
- brands like LG , Vivo , Infinix , Poco, HTC , Google Pixel and IQOO have less than or equal to 3% 

#### Bivariate Analysis

```python
sns.catplot(data=df ,x='Brand',y='Selling Price')
plt.xticks(rotation=45)
```

### Insights
- Apple and samsung are selling phone at ultra preimum category
- followed by Google pixel, motorola , Asus and vivo sellig phone at premium category
- other brands are selling phone at budget category

```python
sns.catplot(data=df, x="Brand", y="Selling Price" ,kind="box")
plt.xticks(rotation=45)
```

### Insights
- Apple and Google Pixel stand out with the highest median selling prices. Apple, in particular, has the largest Interquartile Range (IQR)
- While samsung is concentrated on the budget range but manages with ultra premium segemt too .
- here the dot represents models of a brand that are significantly more expensive than the brand's "usual" offerings.
